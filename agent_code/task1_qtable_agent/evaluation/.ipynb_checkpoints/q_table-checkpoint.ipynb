```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_1 = pd.read_csv('./coin_distance_1/train_log.csv')
df_2 = pd.read_csv('./coin_distance_2/train_log.csv')
df_3 = pd.read_csv('./coin_distance_3/train_log.csv')
df_4 = pd.read_csv('./coin_distance_4/train_log.csv')
df_5 = pd.read_csv('./coin_distance_5/train_log.csv')
display(df_1.head())
```

```python
plt.figure(figsize = (20,10))
plt.plot(df_2['coins_left'],'bo')
plt.plot(df_2['step'], 'ro')
plt.legend(['coins_left','step'])
filename = f"qtable_training_coinheaven_2.png"
plt.savefig(filename)  
plt.show()
```

```python
plt.figure(figsize = (20,10))
#plt.plot(df_1['coins_left'],'bo')
plt.plot(df_1['step'], 'bo')
plt.plot(df_2['step'], 'ro')
plt.plot(df_3['step'], 'ko')
plt.plot(df_4['step'], 'yo')
plt.plot(df_5['step'], 'go')
plt.legend(['coin_distance==1','coin_distance==2','coin_distance==3','coin_distance==4','coin_distance==5'])
filename = f"qtable_training_coinheaven.png"
plt.savefig(filename)  
plt.show()
```
